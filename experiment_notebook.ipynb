```python
from find_max_leaf_binary_tree import mcts_playout
```

```python
import numpy as np
depth = 12
num_iter = 50
num_rollout = 1
# exploration_weight = [0.1, 1, 2, 2.9, 3.5, 4,5,10, 20, 50,80,90, 100, 110, 120,130, 150,200,250,300, 500, 800, 
#                       1000, 5000]
exploration_weight = np.arange(1, 5010, 10)
exploration_weight
```

```python
from collections import defaultdict
num_experiments = 20
exp_res = defaultdict(list)

for c in exploration_weight:
    print("exploration weight:", c)
    for i in range(num_experiments):
        found_max = mcts_playout(depth=depth, num_iter=num_iter, num_rollout=num_rollout, exploration_weight=c)
        exp_res[c].append(found_max)
```

```python
import pandas as pd
exp_res_df = pd.DataFrame(exp_res)
exp_res_df
```

```python
exp_res_df.mean(axis=0).plot(figsize=(30,10),marker='s',color='r', linestyle='solid')
# plt.xticks(exploration_weight);
plt.xlabel("exploration weight")
plt.ylabel("optimal value")
plt.title("results averaged over 20 runs, num_iter = 50, num_rollout = 1")
```

```python
temp = exp_res_df.mean(axis=0)
m = 0
mi = None
for idx, value in zip(temp.index, temp.values):
    if value > m:
        m = value
        mi = idx
print(mi,m)
```

# Grid world RL

```python
import matplotlib.pyplot as plt

from gridworld import GridworldEnv

env = GridworldEnv((9, 9))
env._render(mode="human")
```

```python
from policy_evaluation import mc_policy_evaluation_random_policy
V_sorted = mc_policy_evaluation_random_policy(env,5000)
V_sorted
```

```python

V_plot = np.array([v for state, v in V_sorted])


def plot_gridworld_value(V):
    plt.figure()
    x = np.arange(0.5, 10, 1)  # len = 9
    y = np.arange(0.5, 10, 1)  # len = 9
    c = plt.pcolormesh(x, y,V, cmap="gist_heat")
    plt.colorbar(c)
    plt.gca().invert_yaxis()  # In the array, first row = 0 is on top


# Making a plot always helps
plot_gridworld_value(V_plot.reshape(env.shape))
```
